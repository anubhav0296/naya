```python
# Import required libraries
import seaborn as sns
```

```python
# Distribution plot

# 1. displot (only for 1 feature - histogram) - pdf - kge
# 2. joinplot (x1,x2) - bivariate analysis
# 3. pairplot (more than 2 features) - multivariate
#                                      features


# Categorical plot
# 1. boxplot
# 
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
# Find the correlation matrix - only for numerical data
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
# Join Plot - bivariate analysis (2 feature ke liye hi hota hai)
sns.jointplot(x = 'total_bill', y = 'tip', data= df)
```

```python
# yha hmne hue mai hex use kiya hai jisse hm point ki jagah hexagon bnale
sns.jointplot(x = 'total_bill', y = 'tip', data= df, kind='hex')
```

```python
# 'reg' use krne se wo probability density function aur regression line bnaega
sns.jointplot(x = 'total_bill', y = 'tip', data= df, kind='reg')
```

```python
# Pair plot - isko hm multi-variate analysis ke liye use krte hai
sns.pairplot(df)
```

```python
# hue mai jb sex daalenge toh ye pairplot sex ke aadhar pe karega
sns.pairplot(df, hue= 'sex')
```

```python
# hue mai jb smoker daalenge toh ye pairplot smoker hai ya nahi
# uske aadhar pe karega
sns.pairplot(df, hue= 'smoker')
```

```python
# Dist plot - used for single feature
#             isme kernel density estimation ko true maanke ek curve 
#             bhi print kr dega
sns.distplot(df['tip'])
```

```python
# kyunki isme kde false hai to wo print nahi krega, sirf histogram
# print hoga
sns.distplot(df['tip'], kde= False, bins= 20)
```
